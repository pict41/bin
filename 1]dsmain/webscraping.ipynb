```python
from bs4 import BeautifulSoup
```

```python
import json
```

```python
url="https://www.flipkart.com/apple-iphone-16-black-128-gb/p/itmb07d67f995271?pid=MOBH4DQFG8NKFRDY&lid=LSTMOBH4DQFG8NKFRDYNBDOZI&marketplace=FLIPKART&store=tyy%2F4io&spotlightTagId=default_BestsellerId_tyy%2F4io&srno=b_1_8&otracker=browse&fm=organic&iid=75cf4731-df88-4e3b-9309-c05936435246.MOBH4DQFG8NKFRDY.SEARCH&ppt=browse&ppn=browse&ov_redirect=true"
headers = {
    "User-Agent": "Mozilla/5.0",
    "Content-Type": "application/json"
}
response=requests.get(url,headers=headers)
soup=BeautifulSoup(response.content,"html.parser")
```

```python
scripts=soup.find_all('script',type="application/ld+json")
```

```python
for script in scripts:
    print("Status code:", response.status_code)
    print("Scripts found:", len(scripts))
    try:
        data=json.loads(script.string)
        if isinstance(data, list):
            data = data[0]
        if "review" in data:
            for review in data["review"]:
                print(f"Customer Name:{review["author"]["name"]}")
                print(f"Rating:{review["reviewRating"]["ratingValue"]}")
                print(f"Review:{review["reviewBody"]}")
                print("-" * 60)
    except Exception as e:
        continue
```
